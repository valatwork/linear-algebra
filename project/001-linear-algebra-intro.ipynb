# Introduction to Linear Algebra

If it has an exponential term it *isn't* linear algebra, e.g.: $2x^2 + 5 = 25$ or $2 \sqrt{x} + 5 = 25$.

Linear algebra is used to solve unknowns within a system of linear equations.

In a given system of linear equations, there could be many equations and many unknowns.

There are only three possible outcomes for a system of linear equations:

- No solution
- One solution
- Infinite solutions

#### Example 01 - Linear Equations

```python
import numpy as np
import matplotlib.pyplot as plt
```

- Car A travels at 150 km/h (2.5 km/min) and has a 5 minutes head start
- Car B travels at 180 km/h (3 km/min)
- How long will it take for Car A to catch up with Car B?
- What is the distance travelled by both cars when they meet?

Car A: $d = 2.5t$

Car B: $d = 3(t - 5)$

Car B has a 5 minutes head start, so we need to add that to the time travelled by Car B.

- $2.5t = 3(t - 5)$
- $2.5t = 3t - 15$
- $2.5t - 3t = -15$
- $-0.5t = -15$
- $t = -15 / -0.5$
- $t = 30$ minutes

For the distance travelled by both cars when they meet:

$d = 2.5t = 2.5(30) = 75$ km

```python
t = np.linspace(0, 40, 1000)
car_a = 3 * (t - 5)
car_b = 2.5 * t
fix, axes = plt.subplots()
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in km)')
axes.set_xlim([0, 40])
axes.set_ylim([0, 120])
axes.plot(t, car_a, c='blue')
axes.plot(t, car_b, c='red')
plt.axvline(x=30, color='purple', linestyle='--')
plt.axhline(y=75, color='purple', linestyle='--')
plt.show()
```

#### Example 02 - Linear Equations

Jill designs solar panels as a hobby.

On April 1st, the Mark I design begins generating 1 kj/day
On May 1st, the Mark II design begins generating 4 kj/day

1. On what day will MKII have generated as much energy as MKI?

- $e = 1t$ (for MKI)
- $e = 4(t - 30)$ (for MKII)
- $4(t - 30) = 1t$
- $4t - 120 = t$
- $4t - t = 120$
- $3t = 120$
- $t = 120 / 3$
- $t = 40$ days

2. How much total energy will have both generated by that day?

- $e = 1t = 1(40) = 40$ kj (for MKI)
- $e = 4(t - 30) = 4t - 120$
- $e = 4(t - 30) = 4(40) - 120$
- $e = 160 - 120$
- $e = 40$ kj (for MKII)
- Total energy = 40 + 40 = 80 kj

3. What would be the solutions to 1. and 2. if the Mark II design generated 1 kj/day?

- No solution

```python
t1=np.linspace(0,50,100)

kj_mark_one = t1
kj_mark_two = 4*(t1-30)

fix, axes = plt.subplots()
plt.title('Solar panels')
plt.xlabel('time (in days)')
plt.ylabel('energy (in Kj)')
axes.set_xlim([0,50])
axes.set_ylim([0,90])
axes.plot(t1, kj_mark_one, c='green')
axes.plot(t1, kj_mark_two, c='brown')
plt.axvline(x=40, color='purple', linestyle='--')
plt.axhline(y=40, color='purple', linestyle='--')
```
